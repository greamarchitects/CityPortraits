## Optional: OSM mapping helpers

These cells extend the archival categories into simple OSM-oriented lookups.


```python
osm_mapping = {
    "Courtyard houses": {"building": "residential"},
    "Bazaar fabric": {"landuse": "commercial"},
    "Small plot mixed-use": {"building": "mixed_use"},
    "Commercial strips": {"shop": "*"},
    "Institutional ensembles": {"building": "government"},
    "Perimeter blocks (neoclassical/rationalist)": {"building": "residential"},
    "Monumental civic complexes": {"building": "public"},
    "Palace of Culture": {"building": "civic"},
    "National Historical Museum": {"tourism": "museum"},
    "Prefabricated blocks (4–6 floors)": {"building": "apartments"},
    "Superblocks": {"landuse": "residential"},
    "Slab buildings": {"building": "apartments"},
    "Semi-perimeter blocks": {"building": "apartments"},
    "Freestanding slabs": {"building": "apartments"},
    "Block extensions": {"building": "residential"},
    "Infill buildings": {"building": "residential"},
    "Mixed-use ground floors": {"building": "mixed_use"},
    "Mid/high-rise blocks (>6 floors)": {"building": "apartments"},
    "Worker housing (2–5 floors)": {"building": "apartments"},
    "Textile factories": {"building": "industrial"},
    "Warehouses": {"building": "warehouse"},
    "Industrial halls": {"building": "industrial"},
    "Informal housing clusters": {"building": "residential"},
    "Self-built multi-storey": {"building": "residential"},
    "Incremental housing": {"building": "residential"}
}
```

```python
def map_osm_tags(typologies):
    tags = []
    for t in typologies:
        if t in osm_mapping:
            tags.append(osm_mapping[t])
    return tags

df["OSM Tags"] = df["Building Complexes"].apply(map_osm_tags)
df[["Area / Neighborhood", "Building Complexes", "OSM Tags"]]
```

```python
def format_osm_tags(tags):
    if not tags:
        return "—"
    parts = []
    for tag_dict in tags:
        for k, v in tag_dict.items():
            if v == "*":
                parts.append(f"{k}=*")
            else:
                parts.append(f"{k}={v}")
    return "<br>".join([f"– {p}" for p in parts])

df_osm_display = df.copy()
df_osm_display["Building Complexes"] = df_osm_display["Building Complexes"].apply(to_bullets)
df_osm_display["OSM Tags"] = df_osm_display["OSM Tags"].apply(format_osm_tags)

html = df_osm_display[[
    "Area / Neighborhood",
    "Building Complexes",
    "OSM Tags"
]].to_html(escape=False, index=False)

display(HTML(html))
```

```python
def generate_overpass_query(tags, area="Tirana"):
    queries = []
    for tag_dict in tags:
        for k, v in tag_dict.items():
            if v == "*":
                queries.append(f'node["{k}"]')
                queries.append(f'way["{k}"]')
            else:
                queries.append(f'node["{k}"="{v}"]')
                queries.append(f'way["{k}"="{v}"]')

    query_body = "\n  ".join(queries)
    return f"""
[out:json][timeout:25];
area[\"name\"=\"{area}\"]->.searchArea;
(
  {query_body}
);
out body;
>;
out skel qt;
"""

df["Overpass Query"] = df["OSM Tags"].apply(generate_overpass_query)
df[["Area / Neighborhood", "Overpass Query"]].head(3)
```

```python
df[["Area / Neighborhood", "Overpass Query"]].to_csv(
    "tirana_overpass_queries.csv", index=False
)
```
